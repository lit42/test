```python
import pandas as pd
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
from bokeh.layouts import row
from bokeh.models.widgets import DataTable, TableColumn
```

```python
data = pd.read_csv('1.4_dataset.csv')

```

```python
from bokeh.models import FactorRange

# Create a bar plot for top job titles
top_titles = data['title'].value_counts().head(10)
source_titles = ColumnDataSource(data=dict(title=top_titles.index, count=top_titles.values))

# Manually create a categorical range for x-axis labels
title_range = FactorRange(factors=top_titles.index.tolist())

# Create the figure with the categorical x-axis range
titles_plot = figure(x_range=title_range, title="Top Job Titles", height=300)
titles_plot.vbar(x='title', top='count', width=0.9, source=source_titles)
titles_plot.xaxis.major_label_orientation = 1.2


# Create a bar plot for top platforms
top_platforms = data['platform'].value_counts().head(10)
source_platforms = ColumnDataSource(data=dict(platform=top_platforms.index, count=top_platforms.values))

# Manually create a categorical range for x-axis labels
platform_range = FactorRange(factors=top_platforms.index.tolist())

# Create the figure with the categorical x-axis range
platforms_plot = figure(x_range=platform_range, title="Top Platforms", height=300)
platforms_plot.vbar(x='platform', top='count', width=0.9, source=source_platforms)
platforms_plot.xaxis.major_label_orientation = 1.2


# Create a data table
columns = [
    TableColumn(field="title", title="Job Title"),
    TableColumn(field="location", title="Location"),
    TableColumn(field="salary", title="Salary"),
    TableColumn(field="platform", title="Platform"),
    TableColumn(field="description", title="Description")
]
data_table = DataTable(source=source_titles, columns=columns, width=800, height=300)

# Combine plots and data table
layout = row(titles_plot, platforms_plot, data_table)

output_notebook()
show(layout)
```
